```python
import numpy as np 
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
# Covert equation to a system of first order equations
def system(y, t, k1, k2, m1, m2):
    x1, v1, x2, v2 = y
    dydt = [v1, -1/m1*(k1*x1+k2*(x1-x2)), v2, -1/m2*k2*(x2-x1)]
    return dydt 
```

```python
# Define parameters
k1 = 2
k2 = 2
m1 = 2
m2 = 2
```

```python
# Define initial condition
y0 = [0.1, 0, -0.1, 0]
```

```python
# Define time domain
t = np.linspace(0, 10, 101)
```

```python
# Solve the differential equation
sol = odeint(system, y0, t, args=(k1, k2, m1, m2))
```

```python
# Plot obtained dislacements x1 and x2
plt.plot(t, sol[:,0], label='x1', color='green')
plt.plot(t, sol[:,2], label='x2', color='blue')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
```

```python
# Plot obtained displacement x1 and velocity v1
plt.plot(t, sol[:,0], label='x1', color='green')
plt.plot(t, sol[:,1], label='v2', color='blue')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
```

```python
# Plot obtained velocities v1 and v2
plt.plot(t, sol[:,1], label='v1', color='green')
plt.plot(t, sol[:,3], label='v2', color='blue')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
```
