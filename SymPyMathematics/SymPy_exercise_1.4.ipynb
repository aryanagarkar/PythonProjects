```python
from sympy import *
import numpy as np 
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Intitialize pretty printing
init_printing()

# Variables and functions
l1, m1, l2, m2, t, g = symbols('l1, m1, l2, m2, t, g')
theta1, theta2 = symbols('theta1, theta2', cls=Function)
```

```python
# The kinetic energy
T = 0.5*m1*l1**2*theta1(t).diff(t)**2+0.5*m2*(l1**2*theta1(t).diff(t)**2+l2**2*theta2(t).diff(t)**2+2*l1*l2*theta1(t).diff(t)*theta2(t).diff(t)*cos(theta1(t)-theta2(t)))
```

```python
# The potential energy
V = -g*m1*l1*cos(theta1(t))-g*m2*(l1*cos(theta1(t))+l2*cos(theta2(t)))
```

```python
# The Largrangian equation
L = T - V
```

```python
# The Euler-Lagrange equations
e_l_eq1 = Eq(diff(diff(L, diff(theta1(t), t)), t) - diff(L, theta1(t), 0)
e_l_eq2 = Eq(diff(diff(L, diff(theta2(t), t)), t) - diff(L, theta2(t), 0)
```

```python
omega1, omega2, omegadot1, omegadot2 = symbols('omega1, omega2, omegadot1, omegadot2')
```

```python
# Solve equations
expr1 = Eq(diff(diff(L, diff(theta1(t), t)), t)-diff(L, theta1(t)),0).lhs
expr1 = expr1.subs(Derivative(theta1(t), (t,2)), omegadot1)
expr1 = expr1.subs(Derivative(theta2(t), (t,2)), omegadot2)
expr1 = expr1.subs(Derivative(theta1(t), t), omega1)
expr1 = expr1.subs(Derivative(theta2(t), t), omega2)
```

```python
# Solve equations
expr2 = Eq(diff(diff(L, diff(theta2(t), t)), t)-diff(L, theta2(t)),0).lhs
expr2 = expr2.subs(Derivative(theta1(t), (t,2)), omegadot1)
expr2 = expr2.subs(Derivative(theta2(t), (t,2)), omegadot2)
expr2 = expr2.subs(Derivative(theta1(t), t), omega1)
expr2 = expr2.subs(Derivative(theta2(t), t), omega2)
```

```python
# Solve equations
solve((expr1, expr2),omegadot1,omegadot2)
```
