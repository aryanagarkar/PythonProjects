```python
from sympy import *

# Intitialize pretty printing
init_printing()

# Variables and functions
t, m, k, c = symbols('t, m, k, c')
x = symbols('x', cls=Function)
```

```python
# The kinetic energy of moving mass.
T = 0.5*m*x(t).diff(t)**2
```

```python
# The potential energy of the spring
V = 0.5*k*x(t)**2
```

```python
# The Largrangian L
L = T - V
```

```python
# Addtional factors affecting the response of the system
D = 0.5*c*x(t).diff(t)**2
```

```python
# The Euler-Lagrange equation of motion
e_l_eq = Eq(diff(diff(L, x(t).diff(t)), t) - diff(L, x(t)) + diff(D, x(t).diff(t)), 0)
```

```python
sol = dsolve(e_l_eq)
pprint(sol.rhs)
```

```python
# Variables for plotting solution
C1 = 1
C2 = 1
c = 0.1
k = 10
m = 1
```
