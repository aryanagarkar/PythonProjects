```python
import numpy as np 
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
# Covert equation to a system of first order equations
def pendulum(y, t, l, g):
    theta, omega = y
    dydt = [omega, -g/l * np.sin(theta)]
    return dydt
```

```python
# Define Parameters
l = 1
g = 9.8
```

```python
# Define intial conditions
y0 = [np.pi/2., 0.0]
```

```python
# Define time domain
t = np.linspace(0, 10, 101)
```

```python
# Solve the differential equation
sol = odeint(pendulum, y0, t, args=(l, g))
```

```python
# Plot solution
plt.plot(t, sol[:,0], label='theta', color='green')
plt.plot(t, sol[:,1], label='omega', color='blue')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
```
